# Generate Short URL

Short codes are **5-character random base62 strings**, using the alphabet `0-9`, `a-z`, `A-Z` (62 characters). This gives a space of 62âµ = ~916 million possible codes, when using 5 characters.

```python
import random
import string

BASE62 = string.digits + string.ascii_lowercase + string.ascii_uppercase

def generate_base62(length: int = 5) -> str:
    return "".join(random.choices(BASE62, k=length))
```

```python
generate_base62()
```
